```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('cardiovascular.csv')
group_by_age = df.groupby('age')['cholesterol'].mean().reset_index()
label = group_by_age['age']
value = group_by_age['cholesterol']

fig = plt.subplots(figsize=(19,8))
plt.plot(label, value, label="Average Cholesterol levels per age group", linewidth=3.0)
plt.xlabel('Age')
plt.xticks(rotation='vertical')
plt.ylabel('Cholesterol')
plt.title('Average Cholesterol levels per age group', fontsize=20)
plt.legend()
plt.show()
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('cardiovascular.csv')
group_by_age_systolic = df.groupby('age')['systolic_blood_pressure'].max().reset_index()
systolic_blood_pressure_label = group_by_age_systolic['age']
systolic_blood_pressure_value = group_by_age_systolic['systolic_blood_pressure']
fig = plt.subplots(figsize=(19,8))
plt.plot(systolic_blood_pressure_label, systolic_blood_pressure_value, label="Max systolic blood pressure", linewidth=3.0)

group_by_age_diastolic = df.groupby('age')['diastolic_blood_pressure'].max().reset_index()
group_by_age_diastolic_label = group_by_age_diastolic['age']
group_by_age_diastolic_value = group_by_age_diastolic['diastolic_blood_pressure']
plt.plot(group_by_age_diastolic_label, group_by_age_diastolic_value, label="Max diastolic blood pressure", linewidth=3.0)

plt.xlabel('Age')
plt.xticks(rotation='vertical')
plt.title('A Correlation between systolic and diastolic blood pressure', fontsize=20)
plt.legend()
plt.show()
```
